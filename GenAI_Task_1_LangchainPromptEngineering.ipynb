## Task 1: Replace Hardcoded Prompts

```python
template1 = PromptTemplate(input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template1.format(topic="Machine Learning"))
print(template1.format(topic="Neural Networks"))
```

## Task 2: Multi-Input Prompt System (15%)

```python
template2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print(template2.format(topic="AI", audience="beginners", tone="friendly"))
print(template2.format(topic="Python", audience="kids", tone="fun"))
print(template2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations Engine (15%)

```python
teaching_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} clearly step by step")
interview_template = PromptTemplate(input_variables=["topic"], template="Ask 3 questions about {topic}")
story_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} as a story")
topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System (15%)

```python
role_behaviors = {
    "teacher":     "You are a helpful teacher. Explain things clearly with examples.",
    "interviewer": "You are a professional interviewer. Ask technical questions.",
    "motivator":   "You are an enthusiastic motivator. Inspire the user to learn."
}

def build_chat_prompt(role, topic):
    behavior = role_behaviors[role]
    chat_template = ChatPromptTemplate.from_messages(
        [("system", behavior),("human", "Help me understand: {topic}")])
    return chat_template.format_messages(topic=topic)

messages = build_chat_prompt("teacher", "Neural Networks")
for msg in messages:
    print(f"[{msg.type}]: {msg.content}")
```

## Task 5: Input Validation Layer (10%)

```python
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        print(f"Invalid audience '{audience}', setting to default: 'beginner'")
        audience = "beginner"

    if tone not in valid_tones:
        print(f"Invalid tone '{tone}', setting to default: 'casual'")
        tone = "casual"
    return audience, tone
print(validate_inputs("expert", "formal"))
print(validate_inputs("child", "angry"))
```

## Task 6: Prompt Generator App (15%)

```python
style_templates = {
    "teaching":     PromptTemplate(input_variables=["topic","audience","tone"], template="Explain {topic} for {audience} in a {tone} teaching style"),
    "interview":    PromptTemplate(input_variables=["topic","audience","tone"], template="Generate interview questions on {topic} for {audience} in a {tone} tone"),
    "storytelling": PromptTemplate(input_variables=["topic","audience","tone"], template="Explain {topic} for {audience} in a {tone} storytelling style")
}

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    return style_templates[style].format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print(generate_prompt("Python", "intermediate", "casual", "teaching"))
print(generate_prompt("ML", "expert", "formal", "interview"))
```

## Task 7: Template Reusability Test (10%)

```python
reusable = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [("Blockchain", "beginners", "simple"),("NLP", "researchers", "technical"),("Docker", "developers", "professional"),
    ("Quantum Computing", "students", "exciting"),("Transformers", "engineers", "precise")]

for topic, audience, tone in inputs:
    print(reusable.format(topic=topic, audience=audience, tone=tone))
```
